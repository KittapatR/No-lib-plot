```
#Plotting
class plt:
  
  
  def __init__(self):
    self.plot = ''
    self.sizes = []
  
  
  
  def scatterPlot(self,x, y, rangex = 'default', rangey = 'default',\
                  size = [50,20], xscale = 'default', yscale = 'default'):
    assert [type(x),type(y)] == [list,list], 'Invalid'

    if rangex == 'default': rangex = [min(x),max(x)]
    if rangey == 'default': rangey = [min(y),max(y)]
    if xscale == 'default': xscale = (rangex[1] - rangex[0]) / 5
    if yscale == 'default': yscale = (rangey[1] - rangey[0]) / 5 
      
    minx = min(len(str(xscale)[:str(xscale).index('.')]),2)
    miny = min(len(str(yscale)[:str(yscale).index('.')]),2)

    maxcharx = max( len(str(round(max(x), minx))), len(str(round(abs(min(x)))))) + 3
    maxchary = max( len(str(round(max(y), minx))), len(str(round(abs(min(y)))))) + 3
    divisorx = round( (rangex[1] - rangex[0]) / xscale )
    divisory = round( (rangey[1] - rangey[0]) / yscale )
    
    xnew = [round((i-rangex[0])/(rangex[1]-rangex[0])*size[0])+1 for i in x]
    ynew = [round((i-rangey[0])/(rangey[1]-rangey[0])*size[1])+1 for i in y]
    point = [(i,j) for i, j in zip(xnew, ynew)]
    
    self.plot = ''
    self.sizes = size
    
    for i in range(size[1]+1, -2, -1):
      if i == -1:
        self.plot += ' ' * maxchary
        for j in range(divisorx + 1):
          self.plot += str(round(rangex[0] + j * xscale, minx)) + ' ' * \
          ((size[0]//divisorx) - len(str(round(rangex[0] + j * xscale, minx))))
      else:
        for j in range(0, size[0]+2):
          if i == 0 and j == 0: self.plot += (maxchary + 1) * ' ' + '-'
          elif i == 0: self.plot += '-' 
          elif j == 0 and (i-1) % (size[1] // divisory) == 0: self.plot += (maxchary - \
          len(str(round((i-1)/size[1]*yscale*divisory, miny)+rangey[0]))) * ' '\
            + str(round((i-1)/size[1]*yscale*divisory, miny)+rangey[0] ) + ' |'
          elif j == 0: self.plot += maxchary * ' ' + ' |'
          elif i in ynew:
            if (j, i) in point: self.plot += '*'
            else: self.plot += ' ' 
          else: self.plot += ' '
        self.plot += '\n'
  
  
  
  def hist(self, n, bins = 'default', rangex = 'default', rangey = 'default',\
          xscale = 'default', yscale = 'default', size = [50,20]):
    
    if bins == 'default': it = 1
    else: it = (max(n)-min(n))/(bins-1)
    x = [i for i in range(int(round(min((n)))),int(round(max(n)))+1)]
    y = [0 for i in range(int(round(min((n)))),int(round(max(n)))+1)]
    
    xreduced = list(set(round(i/it) for i in x))
  
    for j in n:
      index = xreduced.index(round(j/it))
      y[index] += 1
      
    xplot = [float(it*i) for i in xreduced]
    
    return self.scatterPlot(xplot,y,rangex = rangex, rangey = rangey, xscale = xscale,\
                           yscale = yscale, size = size)

  
  
  def title(self,name):
    if len(name) < self.sizes[0]: self.plot = (self.sizes[0]//2) * ' ' + \
      str(name) + '\n\n' + self.plot
    else: self.plt = str(name) + '\n\n' + self.plt

  
  
  def ylabel(self,name):
    index = self.plot.find('\n')
    self.plot = self.plot[:index+1] + str(name) + '\n' + self.plot[index+1:]

  
  
  def xlabel(self,name):
    index = self.plot.rfind('\n')
    self.plot = self.plot[:index] + '   ' + str(name) + self.plot[index:]

 

  def show(self):
    return print(self.plot)
```

```
import numpy as np
K = 1000
initial_pop = np.random.normal(100,15,K)

A = plt()
A.hist(initial_pop, size = [100,20])
A.title('Population of bacteria')
A.ylabel('Number')
A.xlabel('Protective score')
A.show()
```
